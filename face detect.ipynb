```python
import cv2
import numpy as np
```

```python
cap= cv2.VideoCapture(0)
face_cascade=cv2.CascadeClassifier(r"C:\Users\Zenbook\Downloads\haarcascade_frontalface_alt.xml")

face_data=[]
dataset_path= "./data"
file_name = input("Enter your name : ")

while True:
    ret, frame = cap.read()
    if ret==False:
        continue
        
    
    faces = face_cascade.detectMultiScale(frame,scaleFactor=1.3,minNeighbors=5)
    
    if len(faces)==0:
        continue
        
    faces = sorted(faces , key = lambda f: f[3]*f[2])   
    
    for face in faces:
        x,y,w,h = face
        cv2.rectangle(frame, (x,y) , (x+w,y+h) , (0,225,225) , 2)
        
        offset=10
        face_section = frame[y-offset:y+h+offset , x-offset:x+w+offset]
        face_section=cv2.resize(face_section , (100,100))
        face_data.append(face_section)
        print(len(face_section))
    
    cv2.imshow("frame",frame)
    
    key_pressed= cv2.waitKey(1) & 0xFF
    if key_pressed== ord('q'):
        break
        
        
face_data= np.asarray(face_data)
face_data=face_data.reshape((face_data.shape[0],-1))
print(face_data.shape)

np.save(dataset_path + file_name + '.npy' , face_data)

print("Data saved!!!")
        
cap.release()
cv2.destroyAllWindows()
```
