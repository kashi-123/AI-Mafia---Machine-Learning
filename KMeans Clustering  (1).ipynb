```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
x,y=make_blobs(n_samples=1000 , n_features=2, centers=5, random_state=6)
print(x.shape)
```

```python
plt.scatter(x[:,0],x[:,1],c=y)
plt.show()
```

```python
k=5
color=['green','yellow','blue','red','gray']
clusters={}
```

```python
for i in range(k):
    center=10*((2*np.random.random((x.shape[1],))-1))
    points=[]
    cluster={
        'center':center,
        'points':points,
        'color':color[i]
    }
    clusters[i]=cluster
```

```python
print(clusters)
```

```python
def distance(v1,v2):
    return np.sqrt(sum((v1-v2)**2))

```

```python
def assign(clusters):
    
    for data_point in range(x.shape[0]):
        dist=[]
        for kx in range(k):
            d=distance(x[data_point],clusters[kx]['center'])
            dist.append(d)
        current_cluster=np.argmin(dist)
        clusters[current_cluster]['points'].append(x[data_point])
```

```python
def updateclusters(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        print(pts.shape)
        if pts.shape[0]>0:
            new_mean=pts.mean(axis=0)
            clusters[kx]['center']=new_mean
            clusters[kx]['points']=[]
            
```

```python
def plotclusters(clusters):
    for kx in range(k):
        pts=np.array(clusters[kx]['points'])
        if pts.shape[0]>0:
            plt.scatter(pts[:,0],pts[:,1],color=clusters[kx]['color'])
        plt.scatter(clusters[kx]['center'][0],clusters[kx]['center'][1],color='black',marker='*')
        
```

```python
assign(clusters)
plotclusters(clusters)
updateclusters(clusters)
```

```python
count =1 
while True:
    assign(clusters)
    updateclusters(clusters)
    count+=1
    if count>1000:
        break
```
